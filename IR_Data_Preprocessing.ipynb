```python
counts = df["review_date"].value_counts() 
counts = counts.sort_index() 

fig, ax = plt.subplots(figsize=(10, 35))
ax.barh(counts.index, counts.values)  
ax.invert_yaxis() 
ax.set_xlabel("Frequency") 
ax.set_ylabel("Date")  
ax.set_title("Review Frequency by Date")  
plt.show()
```

```python
counts = df["report_date"].value_counts()
counts = counts.sort_index() 

fig, ax = plt.subplots(figsize=(10, 35))
ax.barh(counts.index, counts.values)  
ax.invert_yaxis() 
ax.set_xlabel("Frequency") 
ax.set_ylabel("Date")  
ax.set_title("Review Frequency by Report Date")  
plt.show()
```

```python
print(df.groupby('category')['sub_category'].value_counts().sort_index(ascending=True))
```

```python
sns.countplot(x ='helpful_review_count', data = df).set_title("Helpful Review Counts vs Items")
plt.show()
```

```python

label = df['verified_purchase'].value_counts()
plt.pie(label.values,labels=label.index, autopct= '%1.1f%%', startangle=90)
plt.title('Distribution of True Vs False Reviews')
plt.show()
```

```python
sns.catplot(x ='review_rating',kind="count", hue="verified_purchase",data=df)
plt.xlabel("Rating")
plt.ylabel("Number of reviews")
plt.title("Review Rating Grouped by Verified Purchase")
```

```python
sns.catplot(x ='category',kind="count", hue = "verified_purchase", data=df)
plt.xlabel("category")
plt.ylabel("Number of Reviews")
plt.title("Review distribution over Categories")
```

```python
sns.catplot(y ='sub_category',kind="count", hue = "verified_purchase", data=df,  height=5, aspect=3)
plt.ylabel("Sub-Category")
plt.xlabel("Number of Reviews")
plt.title("Review distribution over Sub-Categories")
```

```python
subset = df[["verified_purchase", "review_text"]]

verified = subset[subset["verified_purchase"] == True]
not_verified = subset[subset["verified_purchase"] == False]

verified_avg_length = verified["review_text"].apply(len).mean()
not_verified_avg_length = not_verified["review_text"].apply(len).mean()

x = [verified_avg_length, not_verified_avg_length]
y = ["True", "False"]
sns.barplot(x=x, y=y)
plt.xlabel("Average Length of Reviews")
plt.ylabel("Verified Purchases")
plt.title("Average Length of Reviews Based on Verified Purchases")
plt.show()
```
